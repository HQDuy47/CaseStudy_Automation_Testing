```python
# Đọc file Excel
file_path = 'Testcase.xlsx'
df = pd.read_excel(file_path)

num_rows = len(df)

# Loop through each row
for row_index in range(46):
    column_indices = [3, 4, 5, 6, 7, 8, 9, 10, 11]
    values = [df.iloc[row_index, col] if pd.notna(df.iloc[row_index, col]) else "" for col in column_indices]
    FirstName, MidName, LastName, Email, ContactNo, JobVacacy, Key, Comment, Date = values

    print("FirstName:", FirstName) #3
    print("MidName:", MidName) #4
    print("LastName:", LastName)
    print("Email:", Email)
    print("ContactNo:", ContactNo)
    print("JobVacacy:", JobVacacy)
    print("Key:", Key)
    print("Comment:", Comment)
    print("Date:", Date)

    FirstName_input = driver.find_element("id", "addCandidate_firstName")
    FirstName_input.clear()
    FirstName_input.send_keys(FirstName)

    MidName_input = driver.find_element("id", "addCandidate_middleName")
    MidName_input.clear()
    MidName_input.send_keys(MidName)

    LastName_input = driver.find_element("id", "addCandidate_lastName")
    LastName_input.clear()
    LastName_input.send_keys(LastName)

    Email_input = driver.find_element("id", "addCandidate_email")
    Email_input.clear()
    Email_input.send_keys(Email)

    ContactNo_input = driver.find_element("id", "addCandidate_contactNo")
    ContactNo_input.clear()
    ContactNo_input.send_keys(ContactNo)

    JobVacancy_input = driver.find_element("id", "addCandidate_vacancy")
    JobVacancy_input.send_keys(JobVacacy)


    job_vacancy_select = Select(JobVacancy_input)
    job_vacancy_select.select_by_value("")
    #job_vacancy_select.select_by_value("1")

    Key_input = driver.find_element("id", "addCandidate_keyWords")
    Key_input.clear()
    Key_input.send_keys(Key)

    Comment_input = driver.find_element("id", "addCandidate_comment")
    Comment_input.clear()
    Comment_input.send_keys(Comment)

    formatted_date = Date.strftime("%Y-%m-%d")
    Date_input = driver.find_element("id", "addCandidate_appliedDate")
    Date_input.clear()
    Date_input.send_keys(formatted_date)

    Save_button = driver.find_element("id", "btnSave")
    Save_button.click()

    temp = True
    try:
        wait = WebDriverWait(driver, 1)
        error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//span[@class='validation-error']")))
        error_message = error_message_element.text
        print("Thông báo lỗi:", error_message)
        temp = False
    except TimeoutException:
        temp = True

    if (temp == False):
        print("Bạn thêm không thành công")
         # Open the "Result.xlsx" file
        result_file_path = 'Result.xlsx'
        wb = openpyxl.load_workbook(result_file_path)
        
        # Assuming the "SUCCESSFULLY" column is in the first sheet
        sheet = wb.active
        
        # Get the current row_index
        row_index_in_excel = row_index + 2  # +2 to account for header row and 0-indexing
        
        # Update the "SUCCESSFULLY" column in the corresponding row
        sheet.cell(row=row_index_in_excel, column=14, value="INVALID INPUT")

        # Compare columns 14 and 15 and update the "Result" column accordingly
        successfully_value = sheet.cell(row=row_index_in_excel, column=13).value
        expected_result_value = sheet.cell(row=row_index_in_excel, column=14).value
        result_value = "PASS" if successfully_value == expected_result_value else "FAIL"
        sheet.cell(row=row_index_in_excel, column=15, value=result_value)  # Assuming the "Result" column is in column 16
        
        # Save the changes
        wb.save(result_file_path)

        #delete
        FirstName_input.clear()
        MidName_input.clear()
        LastName_input.clear()
        Email_input.clear()
        ContactNo_input.clear()
        job_vacancy_select.select_by_value("")
        Key_input.clear()
        Comment_input.clear()
        Date_input.clear()

    else: 
        print("Bạn thêm thành công")
        result_file_path = 'Result.xlsx'
        wb = openpyxl.load_workbook(result_file_path)
        
        # Assuming the "SUCCESSFULLY" column is in the first sheet
        sheet = wb.active
        
        # Get the current row_index
        row_index_in_excel = row_index + 2  # +2 to account for header row and 0-indexing
        
        # Update the "SUCCESSFULLY" column in the corresponding row
        sheet.cell(row=row_index_in_excel, column=14, value="SUCCESSFULLY")

        # Compare columns 14 and 15 and update the "Result" column accordingly
        successfully_value = sheet.cell(row=row_index_in_excel, column=13).value
        expected_result_value = sheet.cell(row=row_index_in_excel, column=14).value
        result_value = "PASS" if successfully_value == expected_result_value else "FAIL"
        sheet.cell(row=row_index_in_excel, column=15, value=result_value)  # Assuming the "Result" column is in column 16
        
        # Save the changes
        wb.save(result_file_path)    

        Back_button = driver.find_element("id", "btnBack")
        Back_button.click()

        Add_button = driver.find_element("id", "btnAdd")
        Add_button.click()
```
